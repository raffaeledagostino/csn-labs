# Barabasi Albert

```python
# -----------------------------
# Barabasi albert parameters
# -----------------------------
arrival_times = [1, 10, 100, 1000]
n_runs = 100
n0 = 2
m0 = 2
tmax = 100_000

# BA with preferential attachment
storage_pref, stats_pref, all_deg_pref = run_network_multi(
    n_runs=n_runs, n0=n0, m0=m0, tmax=tmax,
    arrival_times=arrival_times,
    model='BA_pref',
    n_jobs=-1,  
    save_data=True
)


# BA with random attachment
storage_rand, stats_rand, all_deg_rand = run_network_multi(
    n_runs=n_runs, n0=n0, m0=m0, tmax=tmax,
    arrival_times=arrival_times,
    model='BA_random',
    n_jobs=-1,  
    save_data=True
)
```

```python
# ========== BA PREFERENTIAL ==========
fig_raw_pref, fig_rescaled_pref = plot_degree_evolution_separate(
    storage_pref, stats_pref,
    arrival_times,
    n0=n0, m0=m0, tmax=tmax,
    model_name='BA_pref',
    plot_theory=True,      
    plot_rescaled=True,    
    save_dir='./figures'
)

# Model selection
results_pref, fits_pref = model_selection_unweighted(
    storage_pref, stats_pref, arrival_times, tmax=tmax,
    model_name='BA Preferential'
)

# Plot best models
plot_best_models_grid(
    stats_pref, arrival_times, tmax,
    results_pref, fits_pref,
    model_name='BA Preferential',
    save_path='./figures/best_models_BA_pref.png'
)
```

```python

# ========== BA RANDOM  ==========
fig_raw_rand, fig_rescaled_rand = plot_degree_evolution_separate(
    storage_rand, stats_rand,
    arrival_times=arrival_times,
    n0=n0, m0=m0, tmax=tmax,
    model_name='BA_random',
    plot_theory=True,      # Show theoretical curve
    plot_rescaled=True,    # Show data collapse
    save_dir='./figures'
)


results_rand, fits_rand = model_selection_unweighted(
    storage_rand, stats_rand, arrival_times, tmax=tmax,
    model_name='BA Random'
)

plot_best_models_grid(
    stats_rand, arrival_times, tmax,
    results_rand, fits_rand,
    model_name='BA Random',
    save_path='./figures/best_models_BA_random.png'
)
```

# No-growth model

```python
n0_ng = 10_000
m0_ng = 1
tmax_ng = 40 * n0_ng
seed = 42

times, k_traces = no_growth_pa(
    n0=n0_ng, m0=m0_ng, T=tmax_ng, seed=seed
)
save_no_growth_outputs(times, k_traces, prefix="no_growth_pref")

print(k_traces.max())
```

The no-growth model has no arrival times, but the degrees still evolve over time.
Therefore, we save the degree sequence at final time and the time-evolution of 4 chosen vertices (e.g., 0–3) to show “four time-series files per model.” as requested.

```python
plot_no_growth(times, k_traces, n0_ng, m0_ng, t_min=0)
```

```python
k_mean = k_traces.mean(axis=0)
degrees_no_growth = k_traces[:, -1]  # Final degrees at tmax_ng

results, all_fits = model_selection_no_growth(times, k_mean, n0_ng, m0_ng, t_min=n0_ng)

plot_best_model_no_growth(times, k_traces, results, all_fits, show_std=True)
```
